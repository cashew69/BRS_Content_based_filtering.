```python
import pandas as pd
import re
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
stop_words=stopwords.words('english')
```

```python
path = "dataset/books_1.Best_Books_Ever.csv"
```

```python
dataset = pd.read_csv(path)
dataset = dataset[:5000]
dataset[0:5]
```

```python
desc = dataset['description']
```

```python
def remove_stopwords(text):
    words = str(text).split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

```

```python
def clean_text(text):
    # Remove stopwords
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Remove specific characters
    cleaned_words = [re.sub(r"[()':,\"?.]", "", word.lower()) for word in filtered_words]
    return ' '.join(cleaned_words)
```

```python
desc = desc.apply(remove_stopwords)
desc = desc.apply(clean_text)
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vec = cv.fit_transform(desc).toarray()
```

```python
similarity = cosine_similarity(vec)
```

```python
def bookfind(title):
    index = dataset[dataset['title'] == title].index[0]
    distances = similarity[index]
    book_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in book_list:
        print(dataset.iloc[i[0]].title)
```

```python
bookfind('Harry Potter and the Order of the Phoenix')
```
